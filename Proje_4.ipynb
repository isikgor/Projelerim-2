```
import sys
import os
import random as rnd
from google.colab import drive

drive.mount('/content/drive')
os.chdir('/content/drive/My Drive')

class Encryption:
    def __init__(self, upper_base=[], lower_base=[]):
        super().__init__()
        self.upper_base = {}
        self.lower_base = {}

        for i in upper_base:
            self.upper_base[i] = ""

        for j in lower_base:
            self.lower_base[j] = ""

    def load_encrypt_keys(self, upper_key=str, lower_key=str):
        if upper_key == "" or upper_key.isspace():
            raise Exception("üst anahtarda boşluk veya boş olmamalıdır.")

        if lower_key == "" or lower_key.isspace():
            raise Exception("alt anahtarda boşluk veya boş olmamalıdır.")

        upper_keys = upper_key.split(' ')
        lower_keys = lower_key.split(' ')

        for upper in range(0, len(upper_keys)):
            dlist = list(self.upper_base.keys())
            self.upper_base[dlist[upper]] = upper_keys[upper]
        for lower in range(0, len(lower_keys)):
            dlist = list(self.lower_base.keys())
            self.lower_base[dlist[lower]] = lower_keys[lower]
        return True

    def encrypt(self, text):
        result = ""
        for s in range(0, len(text)):
            if text[s] in list(self.upper_base.keys()):
                result = result + str(self.upper_base[text[s]])
            elif text[s] in list(self.lower_base.keys()):
                result = result + str(self.lower_base[text[s]])
            else:
                result = result + str(text[s])
        return result

    def decrypt(self, text):
        result = ""
        for s in text:
            if s in list(self.lower_base.values()):
                for j in self.lower_base:
                    if self.lower_base.get(j) == s:
                        result += j
                        break;
            elif s in list(self.upper_base.values()):
                for j in self.upper_base:
                    if self.upper_base.get(j) == s:
                        result += j
                        break
            else:
                result += s
        return result


class Main:
    def __init__(self):
        super().__init__()

    def read_text_from_file(self, file=str):

        if not os.path.exists(file) or not os.path.isfile(file):
            raise FileNotFoundError()

        reader = open(file, "r", encoding="UTF-8")

        if reader == None:
            raise Exception("Dosya okunamadı.")

        text = reader.read()
        reader.close()
        return text

    def write_text_to_file(self, text=str, file=str):

        if text == "" or text.isspace():
            raise ValueError()

        if os.path.exists(file):
            os.remove(file)

        writer = open(file, "w+")
        writer.write(text)
        writer.close()

    def random_choice(self, text=str, upper_base=[], lower_base=[]):

        result = ""
        for s in text:
            if s in upper_base:
                result = result + \
                    upper_base[rnd.randint(0, len(upper_base) - 1)]
            elif s in lower_base:
                result = result + \
                    lower_base[rnd.randint(0, len(lower_base) - 1)]
            else:
                result = result + s
        return result

    def run(self):

        upper_base = "A B C Ç D E F G Ğ H I İ J K L M N O Ö P R S Ş T U Ü V Y Z"
        lower_base = upper_base.lower().split(' ')
        upper_base = upper_base.split(' ')

        enc = Encryption(upper_base, lower_base)

        menu = "Dosyadaki metni şifrele(E)\n" + \
            "Metnin şifresinin çözüm ekranına yansıt (D)\n" + \
            "Rastgele çözerek şifresini çöz (R)\n" + \
            "Exit (W)\n" + \
            "Lütfen birini seçiniz: "

        choice = str(input(menu))

        if choice == "E":

          try:

            key = str(input("Bir şifreleme anahtarı girin : "))
            lower_key = key.lower()
            upper_key = key.upper()
            enc.load_encrypt_keys(upper_key, lower_key)
            text = self.read_text_from_file(str(input("Bir dosya ekleyin: ")))
            self.write_text_to_file(enc.encrypt(text), str(input("Bir çıktı girin: ")))

          except ValueError:

            print("Bir değer Hatası oluştu.")
            sys.exit(0)

          except FileNotFoundError:

            print("Dosya bulunamadı.")
            sys.exit(0)

          except Exception as ex:
            
            print(ex.message)
            sys.exit(9)

        elif choice == "D":
            key = str(input("Bir Şifreleme Anahtarı Girin: "))
            lower_key = key.lower()
            upper_key = key.upper()

            try:
                enc.load_encrypt_keys(upper_key, lower_key)
                text = self.read_text_from_file(
                    str(input("Bir Dosya ekleyin: ")))
                print(enc.decrypt(text))
            except ValueError:
                print("Bir değer hatası oluştu")
                sys.exit(0)
            except FileNotFoundError:
                print("Dosya Bulunamadı.")
                sys.exit(0)
            except Exception as ex:
                print(ex.message)
                sys.exit(0)

        elif choice == "R":
            try:
                text = self.read_text_from_file(
                    str(input("Bir Dosya ekleyin: ")))
                random_text = self.random_choice(text, upper_base, lower_base)
                print(random_text)
            except FileNotFoundError:
                print("Dosya Bulunamadı.")
                sys.exit(0)
            except Exception as ex:
                print(ex.message)
                sys.exit(0)

        elif choice == "W":
            print("İyi Günler")
            sys.exit(0)

        else:
            print("Faklı Bir Tuşa Bastınız")

main = Main()
main.run()
```
